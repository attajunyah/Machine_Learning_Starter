```python
# map() for Data Frame; returns new, transformed Series and does not modify the original data

review_points_mean = reviews.points.mean()
reviews.points.map(lambda p: p - review_points_mean)
```

```python
# apply() for Data Frame; returns new, transformed DataFrames and does not modify the original data

def remean_points(row):
    row.points = row.points - review_points_mean
    return row

reviews.apply(remean_points, axis='columns')
```

```python
tropical_count = reviews.description.map(lambda desc: "tropical" in desc).sum()
fruity_count = reviews.description.map(lambda desc: "fruity" in desc).sum()
descriptor_counts = pd.Series([tropical_count, fruity_count], index=['tropical', 'fruity'], name = "Counting")
descriptor_counts
```

```python
def rate_points(row):
    if row.country == 'Canada':
        return 3
    elif row.points >= 95:
        return 3
    elif row.points >= 85:
        return 2
    else:
        return 1
    
star_ratings =  reviews.apply(rate_points, axis='columns')

# Check your answer
star_ratings.iloc[:20]
```
