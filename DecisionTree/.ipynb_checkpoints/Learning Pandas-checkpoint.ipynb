```python
# Select the records with index labels 1, 2, 3, 5, and 8, 
# assigning the result to the variable sample_houses
sample_houses = boston_housing.loc[[1, 2, 3, 5, 8], :]
sample_houses
```

```python
# Selecting first 6 rows of 5 different columns 
# With 'loc', we use labels in the table, so you would use the column names 
# So, indexes are inclusively. So 0:10 will select entries 0,...,10
boston_housing.loc[:5, ['crim', 'rm', 'age', 'lstat', 'medv']]
```

```python
# Doing the same with iloc
# Selects first 6 rows of 5 different columns 
# With 'iloc', we use indexes in the table, so you would use the indexes for both row
# and columns. It is similar to python slicing. First element of the range is included
# and the last one excluded. So 0:10 will select entries 0,...,9
boston_housing.iloc[:6, [0, 5, 6, 12, 13]]
```

```python
# This is useful if you can come up with an index
# for the dataset which is better than the current one.
boston_housing.set_index("age")
```

```python
boston_housing.age == 91.0
```

```python
old_houses = boston_housing.loc[(boston_housing.age >= 90) & (boston_housing.age < 100)]
old_houses
```

```python
old_houses = boston_housing.loc[(boston_housing.age >= 90) | (boston_housing.age < 50)]
old_houses
```

```python
boston_housing.loc[boston_housing.indus.isin([7.87, 27.74])]
```

```python
boston_housing.loc[boston_housing.zn.notnull()]
```
